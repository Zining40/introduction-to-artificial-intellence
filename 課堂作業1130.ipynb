```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

# 句子
sentences = [
    "今天天氣真好",
    "不想上課想出去玩",
    "快畢業了耶",
    "糟糕很久沒有讀書了"
]

# 將每個句子分詞
tokenized_sentences = [word_tokenize(sentence) for sentence in sentences]

# 使用Word2Vec模型
word2vec_model = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=5, min_count=1, workers=4)

# 將每個句子轉換為詞向量
sentence_vectors = [word2vec_model.wv[word] for sentence in tokenized_sentences for word in sentence]

# 現在，sentence_vectors 中的每個元素都是相應句子的詞向量
print("Word2Vec Vectors for Sentences:", sentence_vectors)
```

利用Doc2vec

```python
from gensim.models import Doc2Vec
from gensim.models.doc2vec import TaggedDocument
from nltk.tokenize import word_tokenize

# 句子
sentences = [
    "今天天氣真好",
    "不想上課想出去玩",
    "快畢業了耶",
    "糟糕很久沒有讀書了"
]

# 將每個句子分詞並標記
tagged_data = [TaggedDocument(words=word_tokenize(sentence), tags=[str(i)]) for i, sentence in enumerate(sentences)]

# 使用Doc2Vec模型
doc2vec_model = Doc2Vec(tagged_data, vector_size=100, window=2, min_count=1, workers=4)

# 將每個句子轉換為詞向量
sentence_vectors_doc = [doc2vec_model.dv[str(i)] for i in range(len(sentences))]

# 現在，sentence_vectors 中的每個元素都是相應句子的詞向量
print("Doc2Vec Vectors for Sentences:", sentence_vectors_doc)
```

```python
from transformers import BertTokenizer, BertModel
import torch

# 句子
sentences = ["今天天氣真好", "不想上課想出去玩", "快畢業了耶", "糟糕很久沒有讀書了"]

# 使用BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-chinese')
model = BertModel.from_pretrained('bert-base-chinese')

# 轉換成詞向量
sentence_vectors_bert = []
for sentence in sentences:
    inputs = tokenizer(sentence, return_tensors="pt")
    outputs = model(**inputs)
    bert_vector = outputs.last_hidden_state.mean(dim=1).detach().numpy()
    sentence_vectors_bert.append(bert_vector)

# 現在，sentence_vectors 中的每個元素都是相應句子的詞向量
for i, vector in enumerate(sentence_vectors_bert):
    print(f"BERT Vector for Sentence {i + 1}:", vector)
```

比較不同詞向量模型對於兩個句子相似度分數的差異

```python
from gensim.models import Word2Vec, Doc2Vec
from gensim.models.doc2vec import TaggedDocument
from transformers import BertTokenizer, BertModel
from nltk.tokenize import word_tokenize
import torch
from sklearn.metrics.pairwise import cosine_similarity

# 句子
sentences = [
    "今天天氣真好",
    "不想上課想出去玩",
    "快畢業了耶",
    "糟糕很久沒有讀書了"
]

# 使用Word2Vec
word2vec_model = Word2Vec(sentences=[word_tokenize(sentence) for sentence in sentences], vector_size=100, window=5, min_count=1, workers=4)
sentence1_word2vec = [word2vec_model.wv[word] for word in word_tokenize(sentences[0])]
sentence2_word2vec = [word2vec_model.wv[word] for word in word_tokenize(sentences[1])]
similarity_word2vec = cosine_similarity(sentence1_word2vec, sentence2_word2vec)[0][0]

# 使用Doc2Vec
tagged_data = [TaggedDocument(words=word_tokenize(sentence), tags=[str(i)]) for i, sentence in enumerate(sentences)]
doc2vec_model = Doc2Vec(tagged_data, vector_size=100, window=2, min_count=1, workers=4)
sentence1_doc2vec = doc2vec_model.dv[0]
sentence2_doc2vec = doc2vec_model.dv[1]
similarity_doc2vec = cosine_similarity([sentence1_doc2vec], [sentence2_doc2vec])[0][0]

# 使用BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-chinese')
model = BertModel.from_pretrained('bert-base-chinese')

inputs1 = tokenizer(sentences[0], return_tensors="pt")
inputs2 = tokenizer(sentences[1], return_tensors="pt")

outputs1 = model(**inputs1)
outputs2 = model(**inputs2)

bert_vector1 = outputs1.last_hidden_state.mean(dim=1).detach().numpy()
bert_vector2 = outputs2.last_hidden_state.mean(dim=1).detach().numpy()

similarity_bert = cosine_similarity(bert_vector1, bert_vector2)[0][0]

# 打印相似度分數
print("Similarity using Word2Vec:", similarity_word2vec)
print("Similarity using Doc2Vec:", similarity_doc2vec)
print("Similarity using BERT:", similarity_bert)
```
